Importing database using different libraries

1. Manual Function - readline function returns the list that contains all the lines inside the file. comma is encountered split(,) function will be used .
2. Numpy - genfromtxt() is used
3. Pickel file - it saves the file in byte format
4. Pandas -  read_csv()

```python
import numpy as np
import pandas as pd
import pickle
```

```python
def load_csv(filepath):
  data=[]
  col=[]
  checkcol=False
  with open(filepath) as f:
    for val in f.readlines():
      val=val.replace("\n","")
      val=val.split(',')
      if checkcol is False:
        col=val
        checkcol=True
      else:
        data.append(val)
  df=pd.DataFrame(data=data,columns=col)
  return df
print(load_csv('/content/100 Sales Records.csv'))
```

2. Numpy.genfromtxt()

```python
data = np.genfromtxt('/content/100 Sales Records.csv', delimiter = ",")
```

```python
pd.DataFrame(data)
```

to convertnvert Nan into other datatypes

```python
data = np.genfromtxt('/content/100 Sales Records.csv', delimiter = "," , dtype = None)
```

```python
pd.DataFrame(data)
```

Removing b

```python
data = np.genfromtxt('/content/100 Sales Records.csv', delimiter = "," , dtype = None , encoding="utf-8")
```

```python
pd.DataFrame(data)
```

```python
data = np.genfromtxt('/content/100 Sales Records.csv', delimiter = "," , dtype = None , encoding="utf-8", names=True)
```

```python
pd.DataFrame(data)
```
